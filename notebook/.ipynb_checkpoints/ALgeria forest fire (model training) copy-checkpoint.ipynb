```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split 
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_CLEANED_DATASET.csv')
df.head()
```

```python
# dropping the month day and year
df=df.drop(['month','day','year'],axis=1)
```

```python
## assinning values to labels not fire and fire
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
df.head()
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
## Seperating indepndent and dependent features
X = df.drop('FWI',axis=1)
y =df['FWI']
```

```python
y   #dependent features
```

```python
X ## independent features
```

```python
# test trian split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape  ## 182 out of 242 are trianing data set
```

```python
X_test.shape  ## 61 are from test dataset
```

```python
# droping 
```

```python
#Feature Selection on the basis of Coorelation
X_train.corr()
```

```python
##Check for multicolinearity( where the features are highly correlated ex above 95 percent,this type of feature should be remove)
plt.figure(figsize=(12,12))
corr = X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
## Now I want ki jo bhi percentage of correlation meri thershold value se jayada ho usko m remove kardo example 95 per and more
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):        # i want to iterate through each and every element in the heatmap
        for j in range (i):
            if abs(corr_matrix.iloc[i,j]) > threshold:  # Here we are taking abs (absolute value) with which we can be able to manage with  
                colname = corr_matrix.columns[i]        # both positive as well as -ve correlation
                col_corr.add(colname)
    return col_corr
                                                  
```

```python
## Threshold-------> domain expertise(Product Owner decide the value of threshold) 
corr_features=correlation(X_train,0.85)   ## as we can see these two features are highly correlated
```

```python
## droping this to features
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

### Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
## Now after standardization lets check whether we have with the ouliers 
#Box plot
plt.subplots(figsize = (25,5))
plt.subplot(1,2,1)
sns.boxplot(data = X_train)
plt.title('X_train Before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data= X_train_scaled)
plt.title('X_train after scaling')

```

```python
'''AS we can see with Standardization I am done with rectifying the Outliers'''
```

### Linear Regression Model


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled,y_train)
y_predict = lin_reg.predict(X_test_scaled)

mae = mean_absolute_error(y_test,y_predict)

score = r2_score(y_test,y_predict)

print('mean absolute error',mae)
print('R squred value',score)
```

```python
plt.scatter(y_test,y_predict)   ## its a good model but accuracy is very high chances of overfitti
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso   ## as we know it use for feature selection.
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

lasso= Lasso()
lasso.fit(X_train_scaled,y_train)
y_predict = lasso.predict(X_test_scaled)

mae = mean_absolute_error(y_test,y_predict)

score = r2_score(y_test,y_predict)

print('mean absolute error',mae)
print('R squred value',score)
```

```python
plt.scatter(y_test,y_predict)
```

### Cross validation Lasso

```python
from sklearn.linear_model import LassoCV

lasso = LassoCV(cv=5)

lasso.fit(X_train_scaled,y_train)
```

```python
y_pred=lasso.predict(X_test_scaled)
```

```python
lasso.mse_path_
```

```python
lasso.alphas_
```

```python
plt.scatter(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)

score = r2_score(y_test,y_pred)

print('mean absolute error',mae)
print('R squred value',score)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge   ## as we know it use for Overfitting.
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

ridge= Ridge()
ridge.fit(X_train_scaled,y_train)
y_predict = ridge.predict(X_test_scaled)

mae = mean_absolute_error(y_test,y_predict)

score = r2_score(y_test,y_predict)

print('mean absolute error',mae)
print('R squred value',score)
```

```python
plt.scatter(y_test,y_predict)
```

### Ridge CV

```python
from sklearn.linear_model import RidgeCV

ridgecv = RidgeCV(cv=5)  ## here if i provide cv = none then it will experiment on 1 test dataset and other will be training
ridgecv.fit(X_train_scaled,y_train)
y_pred= ridgecv.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)

score = r2_score(y_test,y_pred)
plt.scatter(y_test,y_predict)
print('mean absolute error',mae)
print('R squred value',score)
```

```python
ridgecv.alphas   ## so it consider 10 alphas between 0 and 1 on which it compute or we can say go through the each cv
ridgecv.get_params()
```

### Elastic net (combination of Ridge and Lasso)

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
                                                     ## as we can see with shift + tab elastic has multiple parameters like

                                                        ## L1 , alpha values,etc
elastic_net= ElasticNet()
elastic_net.fit(X_train_scaled,y_train)
y_predict = elastic_net.predict(X_test_scaled)

mae = mean_absolute_error(y_test,y_predict)

score = r2_score(y_test,y_predict)

print('mean absolute error',mae)
print('R squred value',score)
```

```python
plt.scatter(y_test,y_predict)
```

```python
from sklearn.linear_model import ElasticNetCV

elastic_net = ElasticNetCV(cv=5)  ## here in the params we can see it has L1 and L2 norms where it works on 50(l1) - 50(l2) dataset
elastic_net.fit(X_train_scaled,y_train)
y_pred= elastic_net.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)

score = r2_score(y_test,y_pred)
plt.scatter(y_test,y_predict)
print('mean absolute error',mae)
print('R squred value',score)
```
